# Exercise 02.1

## Part 1

```python
n, integral, error = np.loadtxt("ex2.1/integral.txt", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))

#I have plotted a function of number of throws, not blocks
plt.axhline(y=1, color='r', linestyle='-', lw=1, label="Expected")
plt.errorbar(n,integral,yerr=error)
plt.legend()
plt.grid(True)
plt.xlabel("# throws")
plt.ylabel("<integral>" )
plt.show()
```

## Part 2

```python
n, integral, error_importance = np.loadtxt("ex2.1/integral_importance.txt", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))

#I have plotted a function of number of throws, not blocks
plt.axhline(y=1, color='r', linestyle='-', lw=1, label="Expected")
plt.errorbar(n,integral,yerr=error_importance)
plt.legend()
plt.grid(True)
plt.xlabel("# throws")
plt.ylabel("<integral>" )
plt.show()
```

### NOTE
While accuracy of the second extimation is lower (further from the true value) it can be show that the statistical uncertanties are smaller. In the following graph I have plotted the uncertanties from the first method compared to the second.

```python
t=0
plt.figure(figsize=(10,8))
plt.plot(n[t:], error[t:], label="Uniform distribution")
plt.plot(n[t:], error_importance[t:], label="Importance sampling")
plt.grid(True)
plt.xlabel("# throws")
plt.ylabel("Statistical error")
plt.legend()
plt.show()
```

# Exercise 02
## Part 1

```python
n, position, error = np.loadtxt("ex2.2/discrete_lattice.txt", delimiter="\t", unpack=True)
from scipy.optimize import curve_fit

def function(x, k):
    return k*np.sqrt(x)
p_opt, p_cov = curve_fit(function, n, position)
y_fit = function(n,p_opt[0])


plt.figure(figsize=(10,8))

plt.errorbar(n,position,yerr=error, linestyle="", marker=".", label="Simulations")
plt.plot(n,y_fit, label="Fit") # plotting fitted function
plt.legend()
plt.title("Discrete lattice")
plt.grid(True)
plt.xlabel("time step")
plt.ylabel(r"$\sqrt{\langle|r|^2\rangle}$" )
plt.show()
```

## Part 2

```python
n, position, error = np.loadtxt("ex2.2/dense_lattice.txt", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))

p_opt, p_cov = curve_fit(function, n, position)
y_fit = function(n,p_opt[0])

plt.errorbar(n,position,yerr=error, linestyle="", marker=".", label="Simulations")
plt.plot(n,y_fit, label="Fit") # plotting fitted function
plt.legend()
plt.title("Dense lattice")
plt.grid(True)
plt.xlabel("time step")
plt.ylabel(r"$\sqrt{\langle|r|^2\rangle}$" )
plt.show()
```
