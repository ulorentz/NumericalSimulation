```python
import numpy as np
import matplotlib.pylab as plt
```

# Exercise 03.1
## Part 1
Computing option price simulating a single step in time.

```python
n, call_price, error_call, put_price, error_put = np.loadtxt("ex3.1/option_total.txt", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))
plt.errorbar(n,call_price,yerr=error_call)
plt.grid(True)
plt.xlabel("# throws")
plt.ylabel("<call_price>" )
plt.title ("Call Price")

plt.figure(figsize=(10,8))
plt.errorbar(n,put_price,yerr=error_put)
plt.grid(True)
plt.title("Put price")
plt.xlabel("# throws")
plt.ylabel("<put_price>" )
plt.show()

print("Call price: \t",round(call_price[-1],1), "+-",round(error_call[-1],1))
print("Put price: \t",round(put_price[-1],1), "+-",round(error_put[-1],1))
```

## Part 2
Computing option price computing it with 100 steps in time.

```python
n, call_price, error_call, put_price, error_put = np.loadtxt("ex3.1/option_steps.txt", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))
plt.errorbar(n,call_price,yerr=error_call)
plt.grid(True)
plt.xlabel("# throws")
plt.ylabel("<call_price>" )
plt.title ("Call Price")

plt.figure(figsize=(10,8))
plt.errorbar(n,put_price,yerr=error_put)
plt.grid(True)
plt.title("Put price")
plt.xlabel("# throws")
plt.ylabel("<put_price>" )
plt.show()

print("Call price: \t",round(call_price[-1],1), "+-",round(error_call[-1],1))
print("Put price: \t",round(put_price[-1],1), "+-",round(error_put[-1],1))
```
