```python
import numpy as np
import matplotlib.pylab as plt
```

# Exercise 03.1
## Part 1
Option price simulated and computed with a single step in time.

```python
teorical_call=14.975790778311286
teorical_put=5.4595325819072364

n, call_price, error_call, put_price, error_put = np.loadtxt("ex3.1/option_total.txt", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))
plt.axhline(y=teorical_call, color='r', linestyle='-', lw=1, label="Theorical value")
plt.errorbar(n,call_price,yerr=error_call, label= "Simulated")
plt.grid(True)
plt.xlabel("# throws")
plt.legend()
plt.ylabel("<call_price>" )
plt.title ("Call Price")

plt.figure(figsize=(10,8))
plt.axhline(y=teorical_put, color='r', linestyle='-', lw=1, label="Theorical value")
plt.errorbar(n,put_price,yerr=error_put, label="Simulated")
plt.legend()
plt.grid(True)
plt.title("Put price")
plt.xlabel("# throws")
plt.ylabel("<put_price>" )
plt.show()

print("Call price: \t",round(call_price[-1],1), "+-",round(error_call[-1],1))
print("Put price: \t",round(put_price[-1],1), "+-",round(error_put[-1],1))
```

## Part 2
Computing option price with 100 steps in time.

```python
n, call_price, error_call, put_price, error_put = np.loadtxt("ex3.1/option_steps.txt", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))
plt.axhline(y=teorical_call, color='r', linestyle='-', lw=1, label="Theorical value")
plt.errorbar(n,call_price,yerr=error_call, label="Simulated")
plt.grid(True)
plt.legend()
plt.xlabel("# throws")
plt.ylabel("<call_price>" )
plt.title ("Call Price")

plt.figure(figsize=(10,8))
plt.axhline(y=teorical_put, color='r', linestyle='-', lw=1, label="Theorical value")
plt.errorbar(n,put_price,yerr=error_put, label="Simulated")
plt.grid(True)
plt.legend()
plt.title("Put price")
plt.xlabel("# throws")
plt.ylabel("<put_price>" )
plt.show()

print("Call price: \t",round(call_price[-1],1), "+-",round(error_call[-1],1))
print("Put price: \t",round(put_price[-1],1), "+-",round(error_put[-1],1))
```
