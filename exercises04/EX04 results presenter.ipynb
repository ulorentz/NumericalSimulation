This notebook simply prints and plots the outputs generated with `ex4.4/*/MolDyn`. To compile and run, simply do, in the directory of this notebook:

`make && ./run.sh`

```python
import numpy as np
import matplotlib.pylab as plt

#functions definitions
def plot_temperature(time, temp, temp_error, title):
    plt.figure(figsize=(10,8))
    plt.errorbar(time,temp,yerr=temp_error)
    plt.grid(True)
    plt.xlabel("time [ns]")
    plt.ylabel("temperature [k]" )
    plt.title (title)
    plt.show()
    
def plot_energy(time, energy, energy_error, kin, kin_error, pot, pot_error, title):
    plt.figure(figsize=(10,8))
    plt.errorbar(time,energy,yerr=energy_error, label="Total Energy")
    plt.errorbar(time,kin,yerr=kin_error, label="Kinetic Energy")
    plt.errorbar(time,pot,yerr=pot_error, label="Potential Energy")
    plt.legend()
    plt.grid(True)
    plt.xlabel("time [ns]")
    plt.ylabel("Energy [J]" )
    plt.title (title)
    plt.show()

def plot_pressure(time, press, press_error, title):
    plt.figure(figsize=(10,8))
    plt.errorbar(time,press,yerr=press_error)
    plt.grid(True)
    plt.xlabel("time [ns]")
    plt.ylabel("Pressure [Pa]" )
    plt.title (title)
    plt.show()

# Read data
path="ex4.4/solid/results/"
temp_solid, temp_error_solid = np.loadtxt(path+"ave_temp.out", delimiter="\t", unpack=True)
energy_solid, energy_error_solid = np.loadtxt(path+"ave_etot.out", delimiter="\t", unpack=True)
kin_solid, kin_error_solid = np.loadtxt(path+"ave_ekin.out", delimiter="\t", unpack=True)
pot_solid, pot_error_solid = np.loadtxt(path+"ave_epot.out", delimiter="\t", unpack=True)
press_solid, press_error_solid = np.loadtxt(path+"ave_press.out", delimiter="\t", unpack=True)

path="ex4.4/gas/results/"
temp_gas, temp_error_gas = np.loadtxt(path+"ave_temp.out", delimiter="\t", unpack=True)
energy_gas, energy_error_gas = np.loadtxt(path+"ave_etot.out", delimiter="\t", unpack=True)
kin_gas, kin_error_gas = np.loadtxt(path+"ave_ekin.out", delimiter="\t", unpack=True)
pot_gas, pot_error_gas = np.loadtxt(path+"ave_epot.out", delimiter="\t", unpack=True)
press_gas, press_error_gas = np.loadtxt(path+"ave_press.out", delimiter="\t", unpack=True)

path="ex4.4/liquid/results/"
temp_liquid, temp_error_liquid = np.loadtxt(path+"ave_temp.out", delimiter="\t", unpack=True)
energy_liquid, energy_error_liquid = np.loadtxt(path+"ave_etot.out", delimiter="\t", unpack=True)
kin_liquid, kin_error_liquid = np.loadtxt(path+"ave_ekin.out", delimiter="\t", unpack=True)
pot_liquid, pot_error_liquid = np.loadtxt(path+"ave_epot.out", delimiter="\t", unpack=True)
press_liquid, press_error_liquid = np.loadtxt(path+"ave_press.out", delimiter="\t", unpack=True)

#Setting time
moves_block=10000/50
time=np.arange(temp_solid.size)*moves_block/10 #nanosecond
```

# Exercise 04.4

# Argon

```python
#parameters for argon
sigma = 0.34    
epsilon_boltzman = 120  #e/kB
k_boltzman=1.38064852*1E-23
epsilon=epsilon_boltzman*k_boltzman
sigma_si=sigma*1E-9
mass = 39.948*1.66054*1E-27
```

## Solid

```python
title="Solid Argon"

#temperature
temp=epsilon_boltzman*temp_solid
temp_error=epsilon_boltzman*temp_error_solid
plot_temperature(time, temp, temp_error, title)

#energy
kin=kin_solid*epsilon
kin_error=kin_error_solid*epsilon
pot=pot_solid*epsilon
pot_error=pot_error_solid*epsilon
energy=energy_solid*epsilon
energy_error=energy_error_solid*epsilon
plot_energy(time, energy, energy_error, kin, kin_error, pot, pot_error, title)

#pressure
press=press_solid*epsilon/(sigma_si**3)
press_error=press_error_solid*epsilon/(sigma_si**3)
plot_pressure(time, press, press_error, title)

```

## Liquid

```python
title="Liquid Argon"

#temperature
temp=epsilon_boltzman*temp_liquid
temp_error=epsilon_boltzman*temp_error_liquid
plot_temperature(time, temp, temp_error, title)

#energy
kin=kin_liquid*epsilon
kin_error=kin_error_liquid*epsilon
pot=pot_liquid*epsilon
pot_error=pot_error_liquid*epsilon
energy=energy_liquid*epsilon
energy_error=energy_error_liquid*epsilon
plot_energy(time, energy, energy_error, kin, kin_error, pot, pot_error, title)

#pressure
press=press_liquid*epsilon/(sigma_si**3)
press_error=press_error_liquid*epsilon/(sigma_si**3)
plot_pressure(time, press, press_error, title)
```

## Gas

```python
title="Gas Argon"

#temperature
temp=epsilon_boltzman*temp_gas
temp_error=epsilon_boltzman*temp_error_gas
plot_temperature(time, temp, temp_error, title)

#energy
kin=kin_gas*epsilon
kin_error=kin_error_gas*epsilon
pot=pot_gas*epsilon
pot_error=pot_error_gas*epsilon
energy=energy_gas*epsilon
energy_error=energy_error_gas*epsilon
plot_energy(time, energy, energy_error, kin, kin_error, pot, pot_error, title)

#pressure
press=press_gas*epsilon/(sigma_si**3)
press_error=press_error_gas*epsilon/(sigma_si**3)
plot_pressure(time, press, press_error, title)
```
