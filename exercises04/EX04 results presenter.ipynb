## Ex 04.2

```python
temp, temp_error = np.loadtxt("ex4/results/ave_temp.out", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))
plt.errorbar(np.arange(temp.size),temp,yerr=temp_error)
plt.grid(True)
plt.xlabel("time")
plt.ylabel("temperature" )
plt.title ("Temperature")
plt.show()

energy, energy_error = np.loadtxt("ex4/results/ave_etot.out", delimiter="\t", unpack=True)

kin, kin_error = np.loadtxt("ex4/results/ave_ekin.out", delimiter="\t", unpack=True)

pot, pot_error = np.loadtxt("ex4/results/ave_epot.out", delimiter="\t", unpack=True)

plt.figure(figsize=(10,8))
plt.errorbar(np.arange(energy.size),energy,yerr=energy_error, label="Total Energy")
plt.errorbar(np.arange(kin.size),kin,yerr=kin_error, label="Kinetic Energy")
plt.errorbar(np.arange(pot.size),pot,yerr=pot_error, label="Potential Energy")
plt.legend()
plt.grid(True)
plt.xlabel("time")
plt.ylabel("Energy" )
plt.title ("Energy")
plt.show()

press, press_error = np.loadtxt("ex4/results/ave_press.out", delimiter="\t", unpack=True)
plt.figure(figsize=(10,8))
plt.errorbar(np.arange(press.size),press,yerr=press_error)
plt.grid(True)
plt.xlabel("time")
plt.ylabel("Pressione" )
plt.title ("Pressione")
plt.show()
```
