```python
import numpy as np                                                                 
import matplotlib.pyplot as plt    
%matplotlib inline
```

# Genetic Algorithm
The genetic algorithm I have implemented use the following strathegy.
During each iteration some kind of genetic transformation is done to the whole population but the best two elements, that are saved and kept for the next population. To the others a crossover is performed with 70% of probability, if that is not the case one of the five possibile mutation takes place with equal probability. At the end of one iteration there is a new population, with the two best elements of the previous and all the others changed. 

The algorithm for finding the convergence is very naive and actually performs more iteration than necessary. 

# Circular 30 cities

```python
x, y=np.loadtxt("outputs/circle_cities_location.dat", delimiter="\t", unpack=True)                            
fig,ax = plt.subplots(2, 2, figsize=(16, 16))

ax[0][0].plot(x, y, marker=".", linestyle="");
ax[0][0].set_title("Cities locations");

initial=np.loadtxt("outputs/circle_initial_best_path.dat",dtype="uint32")  
x_initial=x[initial]
y_initial=y[initial]
ax[0][1].plot(x_initial, y_initial, marker="."); 
ax[0][1].set_title("Initial best path");

final=np.loadtxt("outputs/circle_L1_final_best_path.dat",dtype="uint32")  
x_final=x[final]
x_final=np.append(x_final, x[final[0]])
y_final=y[final]
y_final=np.append(y_final, y[final[0]])
ax[1][0].plot(x_final, y_final, marker=".");  
ax[1][0].set_title("Final L1 best path");

final=np.loadtxt("outputs/circle_L2_final_best_path.dat",dtype="uint32")  
x_final=x[final]
x_final=np.append(x_final, x[final[0]])
y_final=y[final]
y_final=np.append(y_final, y[final[0]])
ax[1][1].plot(x_final, y_final, marker=".");  
ax[1][1].set_title("Final L2 best path");
```

```python

ave=np.loadtxt("outputs/fit_av_circle_L1.dat") 
best=np.loadtxt("outputs/fit_best_circle_L1.dat") 

fig,ax = plt.subplots(1, 2, figsize=(16, 6))

ax[0].loglog(ave, label="Averaged")
ax[0].loglog(best, label="Best")
ax[0].set_title("L1")
ax[0].legend()
ave=np.loadtxt("outputs/fit_av_circle_L2.dat") 
best=np.loadtxt("outputs/fit_best_circle_L2.dat") 
ax[1].loglog(ave, label="Averaged")
ax[1].loglog(best, label="Best")
ax[1].set_title("L2")
ax[1].legend();
```
