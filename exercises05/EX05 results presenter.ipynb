## Instructions  
This notebook simply prints and plots the outputs generated by `./hydro`. To compile and run, simply do, in the directory of this notebook:

`make && ./hydro`

Note that, if you want to visualize the distribution of points, you can run `./hydro --pos` and then execute the python script `visualizer.py` in this folder. With the previous option, the program will print the positions and not the mean radius (without the option it behaves in the opposite way). Due to the default high number of points ($10^6$) the visualizer may be very slow. In order to reduce the number of points, a second argument may be given to the program: `./hydro --pos n`, where n is an integer. Furthermore you can set the way the step are generated. For info type `./hydro --help`.

```python
import numpy as np
import matplotlib.pylab as plt
```

## Probability density 100
### Uniformly distributed 

```python
r, error = np.loadtxt("radius100uniform.txt", delimiter="\t", unpack=True)

plt.figure(figsize=(10,8))
plt.errorbar(np.arange(r.size),r,yerr=error)
plt.grid(True)
plt.xlabel("Block")
plt.ylabel("radius [bohr units]" )
plt.title ("Radius 100 uniform")
plt.show()
```

```python
print("Error on the last block: ", error[-1])
```

### Gaussian distributed

```python
r, error = np.loadtxt("radius100gauss.txt", delimiter="\t", unpack=True)

plt.figure(figsize=(10,8))
plt.errorbar(np.arange(r.size),r,yerr=error)
plt.grid(True)
plt.xlabel("Block")
plt.ylabel("radius [bohr units]" )
plt.title ("Radius 100 gaussian")
plt.show()
```

```python
print("Error on the last block: ", error[-1])
```
